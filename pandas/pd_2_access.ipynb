# iloc 과 loc

## iloc vs loc

iloc은 숫자 기반으로 dataframe에 접근 하고 싶을때 사용한다.

반대로 loc은 라벨(이름) 기반으로 접근 하고 싶을때 사용한다.

---
    df.iloc[]은 해당 row의 값들에 접근한다.

```python
ex_data.iloc[0]
```

    df.iloc[]은 슬라이싱이 가능하다.

```python
ex_data.iloc[:,0]
```

```python
ex_data.iloc[-5:]
```

```python
ex_data.loc[9 , "senior_population"]
```

```python
ex_data.loc[:, ["senior_population", "children_population"]]
```

# many accessor method

    (1) 특정 열과 해당 값을 논리연산이 가능하다.

```python
ex_data.senior_population == 5957
```

    (2) loc과 논리연산자를 통해 특정 값의 행에 접근 가능하다.

```python
ex_data.loc[ex_data.senior_population == 5957]
#ex_data.loc["senior_population" == 5957]
```

```python
ex_data.loc[(ex_data.senior_population == 5957) & (ex_data.children_population >= 649)]
```

```python
ex_data.loc[(ex_data.senior_population == 5957) | (ex_data.children_population >= 649)]
```

    (3) isin을 통해 해당 값이 속한 행을 추출할 수 있다.

```python
ex_data.loc[ex_data.senior_population.isin([5957, 4665])]
```

    (4) notnull()을 통해 비어있지 않은 DataFrame을 추출할 수 있다.

```python
ex_data.loc[ex_data.senior_population.notnull()]
```

    (5) 특정 열의 value들을 바꿀 수 있다.

```python
ex_data['children_population'] = 'child' 
ex_data['children_population']
```
